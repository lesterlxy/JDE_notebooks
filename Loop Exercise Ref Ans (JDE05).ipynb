### Exercise 3

#### Write a program in Python to produce Star triangle as shown below.


Note: 
- Total number of lines is 9

- Number of * is 1 at first line
- Number of * is 17 at last line
- Increment of * is 2

- Pattern of * -> 2*x-1 if x starts from 1 to 9, inclusive
- Pattern of *-> 2*x+1 if x starts from 0 to 8, inclusive

- Number of spaces is 8 at first line
- Number of spaces is 0 at last line
- Decrement of spaces is 1

- Pattern of space -> 9-x if x starts from 1 to 9, inclusive 
    - If we let r be number of lines, then it would be r-x

- Pattern of space -> 8-x if x starts from 0 to 8, inclusive
    - If we let r be number of lines, then it would be (r-1)-x

```python
#creating a function called pyfunc

def pyfunc(r):
    for x in range(r):   # x will go from 0 to 9, exclusive 9
        print(' '*(r-1-x) + '*'*(2*x+1) )
        
#calling the function and specifying the argument to be 9
pyfunc(9)
```
