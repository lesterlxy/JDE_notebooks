**Create a user defined function to check the validity of an NRIC number.**


```python
def checkNRIC(nric: str) -> bool:
    weights = [2, 7, 6, 5, 4, 3, 2]
    check_char_list = ["A", "B", "C", "D", "E", "F", "G", "H", "I", "Z", "J"]
    csum = 0
    for i in range(7):
        csum = csum + int(nric[i + 1]) * weights[i]
    csum = csum % 11
    csum = 11 - csum
    check_char = check_char_list[csum - 1]
    if nric[8].upper() == check_char:  # use .upper() to handle lowercase
        return True
    else:
        return False
```

```python
display(checkNRIC("S1234567A"))  # False
display(checkNRIC("S9945106H"))  # True
display(checkNRIC("S1105876J"))  # True
```
