### Create a user defined function to check the validity of an NRIC number.

#### 1) Checking for S prefix only

```python
def nric_checker(nric):
    
    weight = (2,7,6,5,4,3,2)
    reference = ('A','B','C','D','E','F','G','H','I','Z','J')
    
    ref = nric[-1] 
    sum_total = 0
    
    nric_digit=nric[1:8]
    
    for i in range(len(nric_digit)):
        sum_total = sum_total + int(nric_digit[i])*weight[i]
        
    ref_index = 11 - sum_total%11
    
    if ref == reference[ref_index - 1]:
        return True
    return False
```

```python
# Can use S1234567A which is not correct to test, S1234567D would be correct.

nric_number= input("What is your NRIC number, including S?").strip()
nric_checker(nric_number)
```

#### 2) Checking for S and T prefix

```python
def nric_checker(nric):
    
    weight = (2,7,6,5,4,3,2)
    
    if nric[0] == 'S':
        reference = ('A','B','C','D','E','F','G','H','I','Z','J')
    else:
        reference = ('H','I','Z','J','A','B','C','D','E','F','G')
        
    ref = nric[-1] 
    sum_total = 0
    
    nric_digit=nric[1:8]
    
    for i in range(len(nric_digit)):
        sum_total = sum_total + int(nric_digit[i])*weight[i]
        
    ref_index = 11 - sum_total%11
    
    if ref == reference[ref_index - 1]:
        return True
    return False
```

```python
# Can additionally use T1234567A which is not correct to test, T1234567J would be correct.
nric_number= input("What is your NRIC number, including S/T?").upper().strip()
nric_checker(nric_number)
```

#### 3) Checking for S, T, F and G prefix

```python
def nric_checker(nric):
    
    weight = (2,7,6,5,4,3,2)
    
    if nric[0] == 'S':
        reference = ('A','B','C','D','E','F','G','H','I','Z','J')
    elif nric[0] == 'T':
        reference = ('H','I','Z','J','A','B','C','D','E','F','G')
    elif nric[0] == 'F':
        reference = ('K','L','M','N','P','Q','R','T','U','W','X')
    else:
        reference = ('T','U','W','X','K','L','M','N','P','Q','R')
        
    ref = nric[-1] 
    sum_total = 0
    
    nric_digit=nric[1:8]
    
    for i in range(len(nric_digit)):
        sum_total = sum_total + int(nric_digit[i])*weight[i]
        
    ref_index = 11 - sum_total%11
    
    if ref == reference[ref_index - 1]:
        return True
    return False
```

```python
# Can additionally use F1234567A which is not correct to test, F1234567N would be correct.
nric_number= input("What is your NRIC number, including S/T/F/G?").upper().strip()
nric_checker(nric_number)
```

```python
def nric_checker(nric):
    
    weight = (2,7,6,5,4,3,2)
    
    if nric[0] in 'ST':
        reference = ('A','B','C','D','E','F','G','H','I','Z','J')
    else:
        reference = ('K','L','M','N','P','Q','R','T','U','W','X')

        
    ref = nric[-1] 
    sum_total = 0
    
    nric_digit=nric[1:8]
    
    for i in range(len(nric_digit)):
        sum_total = sum_total + int(nric_digit[i])*weight[i]
    
    if nric[0] in 'TG':
        sum_total = sum_total + 4
        
    ref_index = 11 - sum_total%11
    
    if ref == reference[ref_index - 1]:
        return True
    return False
```

```python
# Can additionally use F1234567A which is not correct to test, F1234567N would be correct.
nric_number= input("What is your NRIC number, including S/T/F/G?").upper().strip()
nric_checker(nric_number)
```
